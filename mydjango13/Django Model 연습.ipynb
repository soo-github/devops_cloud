# Conditions

```python
qs = Shop.objects.all()
```

```python
# fillter <-> exclude
qs.filter(pk=1)
```

```python
qs.exclude(pk=1)
```

```python
# lt: less than               <
# lte: less than equal        <=
# gt: greater than            >
# gte: greater than equal     >=
qs.filter(
    pk__gte=5,
    pk__lt=10, # less than
)
```

```python
qs.filter(pk__gte=5).filter(pk__lt=10)
```
